Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
In Bokeh, glyphs are visual shapes or markers that are used to represent data points on a plot. Bokeh provides a variety of glyph types, such as circles, squares, lines, and many others, that can be customized in terms of size, color, transparency, and other properties.

To add glyphs to a Bokeh plot, you can use the figure() method to create a plot object, and then call a glyph method, such as circle(), square(), or line(), to add the desired glyphs to the plot.

Here's an example of how to create a scatter plot with circles as glyphs using Bokeh:

from bokeh.plotting import figure, output_file, show

# create a plot with specified size and title
plot = figure(plot_width=400, plot_height=400, title='Scatter Plot')

# create some sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# add circles as glyphs to the plot
plot.circle(x, y, size=10, color='navy', alpha=0.5)

# specify output file and show the plot
output_file('scatter.html')
show(plot)

Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?

Bokeh server makes it easy to create interactive web applications that connect front-end UI events to running Python code. Bokeh creates high-level Python models, such as plots, ranges, axes, and glyphs, and then converts these objects to JSON to pass them to its client library, BokehJS.

The Bokeh server allows you to run a Python script that creates a plot, and then serve that plot to one or more web browsers. The plot can be updated in real time based on user input or external data sources. Here are the general steps to create an interactive plot using the Bokeh server.

Create a Bokeh plot in a Python script, using the bokeh.plotting module.
Define a function to update the plot, which takes in any necessary parameters and updates the plot data or configuration.
Create a Bokeh server application, using the bokeh.application module, which defines the structure of the application and how the plot will be updated.
Run the Bokeh server, which serves the application to one or more web browsers, using the bokeh.server module.

<span id="userguide-embed-standalone-autoload"></span><h3>Autoloading scripts<a class="headerlink" href="#autoloading-scripts" title="Permalink to this heading">#</a></h3>
<p>You can also embed standalone documents with the <a class="reference internal" href="../reference/embed.html#bokeh.embed.autoload_static" title="bokeh.embed.autoload_static"><code class="xref py py-func docutils literal notranslate"><span class="pre">autoload_static()</span></code></a> function.
This function provides a <code class="docutils literal notranslate"><span class="pre">&lt;script&gt;</span></code> tag that replaces itself with a Bokeh
plot. This script also checks for BokehJS and loads it if necessary. This
function lets you embed a plot with nothing but this <code class="docutils literal notranslate"><span class="pre">&lt;script&gt;</span></code> tag.</p>
<p>This function takes a Bokeh model, such as a plot, that you want to display, a
<code class="docutils literal notranslate"><span class="pre">Resources</span></code> object, and a path to load a script from. Then <a class="reference internal" href="../reference/embed.html#bokeh.embed.autoload_static" title="bokeh.embed.autoload_static"><code class="xref py py-func docutils literal notranslate"><span class="pre">autoload_static()</span></code></a>
returns a self-contained <code class="docutils literal notranslate"><span class="pre">&lt;script&gt;</span></code> tag and a block of JavaScript code. The
JavaScript code saves to the path you provide and the <code class="docutils literal notranslate"><span class="pre">&lt;script&gt;</span></code> loads and
runs it to display your plot on a web page.</p>

<span id="userguide-embed-apps"></span><h2>Bokeh applications<a class="headerlink" href="#bokeh-applications" title="Permalink to this heading">#</a></h2>

embed Bokeh apps so that every page load either creates and displays a new

<section id="app-documents">

<section id="app-sessions">
<h3>App sessions<a class="headerlink" href="#app-sessions" title="Permalink to this heading">#</a></h3>
<p>Sometimes, instead of loading a new session, you might wish to load a
<em>specific</em> one.</p>
<p>Take a Flask app that renders a page for an authenticated user. You might want
it to pull a new session, make some customizations for that specific user, and
serve this customized Bokeh server session.</p>
<p>You can accomplish this with the <a class="reference internal" href="../reference/embed.html#bokeh.embed.server_session" title="bokeh.embed.server_session"><code class="xref py py-func docutils literal notranslate"><span class="pre">server_session()</span></code></a> function. This function
accepts a specific model to embed (or <code class="docutils literal notranslate"><span class="pre">None</span></code> for an entire session document),
session ID, and a URL to the Bokeh application.</p>
